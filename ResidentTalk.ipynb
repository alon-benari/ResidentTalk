# Measure of dispersion and location  or is there a way to capture the who distribution in a short succinct way?

One of the most common mesure for location and dispersion are the mean  and standard deviation respectively. The mean is also know as the expected value or expectancy and is denoted as $\mu$ or $E(x)$. Standard deviation


$$\mu = E(x) = \frac{\sum_{i=1}^{N}(x_i)}N$$

Standard deviation is defined as 

$$\mu = \sqrt{E(x^2)-E(x)^2} $$

Lets calculates the mean and standard deviation for our sample

```python
mu, sigma = np.mean(chips),np.std(chips)
print 'mean is: %4.3f' %mu
print 'standard deviation is %4.3f' %sigma
```

Lets plot those results and see how that looks like:

```python
fig, ax = plt.subplots(1,1)
fig.set_figwidth(8)
fig.set_figheight(10)
ax.hist(chips,bins =b0 ,normed=0,alpha=0.4)
ax.set_ylabel('count',fontsize=12)
ax.axvline(mu,linestyle=':')
ax.axvline(mu+sigma,linestyle='--')
ax.axvline(mu-sigma,linestyle='--')
ax.axvline(mu+1.96*sigma,linestyle='-.')
ax.axvline(mu-1.96*sigma,linestyle='-.')
```

What we observe is that the mean and the standard deviation are good at capturing the dispersion or variance of the sample at hand.
Lets look at another even more fundamental measure called median and inter-quartile range or IQR. This is basically looking at the quantiles. For example, the 25% quantile answers the question of what is the observation which 25% of the disctibution is below?
Median is defined as the value which 50% of the sample is greater/lesser from. The IQR are the values in the sample  25% of which the are above and below 75%  i.e. representing the 25-75% quantile range.
In our example  be get:

## Why two measures? and the uniquity of the normal distribution.

If one was good than why do I need two?. Turns out you probably do not need two, but they have their advantages.
Mean and standard deviation are good measure for a specific distribution that is known as the normal distribution. Because of its ubiquity lets take a minute to look closely at this probability distribution in light of what we learned from the measures of location and dispersion. Lets look at a few examples from the normal distribution parametrized by the mean and the standard deviation.


$$P(X|\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}exp^{\frac{-(x-\sigma)^2}{\sigma^2}}$$

```python
%matplotlib inline
random.seed( 10 )
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)

##measures of dispersion
def dispersion(mu,s1,s2,s3):
    sigma = [s1,s2,s3]  # make list of stds
    col = ['crimson', 'blue', 'chartreuse']  
    
    fig,ax = plt.subplots()
    fig.set_figwidth(12)
    fig.set_figheight(8)
    ax.set_title('Histograms of different dispersions',fontsize=12)
    ax.set_ylabel('count',fontsize=12)
    ax.set_xlabel('quantile',fontsize=12)
    for i,s in enumerate(sigma):
        l1 = 'sigma='+str(s)
        sample = np.random.normal(mu,s,1000)
        b=-0.5+np.arange(round(min(sample)),round(max(sample)))
        h,b = np.histogram(sample,bins = b)
        ax.hist(sample,
                bins=b,
                color=col[i], normed=0,
                alpha=0.4,label=l1)
        ax.plot(round(b[0])+np.diff(b).cumsum(),h,color=col[i])
    ax.legend()
    fig.show()
dispersion(2,7,5,2) # 
```

we observe that the location and scale coefficients will determine the shape of the  distribtion and its location.
Lets see what are the median mean and standard deviation.

```python
random.seed( 10 )
sample = np.random.normal(0,5,100000) # sample 100,000 from a Normal with mean 0 and variance of 5
mu, sigma = np.mean(sample) , np.std(sample)
low = mu-2*sigma
up = mu+2*sigma
#
q = mquantiles(sample,[0.025,0.5,0.975]) # compute quantiles. 2.5%, 50% and 97.5%

print "The mean is %4.2f, stdev is %4.2f" %(mu,sigma)
print "The upper and lower bounds (+/- 2*stdev) are %4.2f, %4.2f" %(low,up)
print "The mode and 95 percent quantile are %4.2f,%4.2f , %4.2f" %(q[0],q[1],q[2])


```

So we observe something interesing here: The upper and lower border which are 2 stds from the mean closely resemble the 2.5% and 97.5 
quantiles i.e. 95% of the distribution is between these  limits. The median corresponds roughly to the mean. Lets see how this look graphicaly. 

```python
random.seed( 10 )
sample = np.random.normal(mu,sigma,100000)
b=-0.5+np.arange(round(min(sample)),round(max(sample)))
h,b = np.histogram(sample,bins = b,normed=1)
fig,ax = plt.subplots(1)
fig.set_figwidth(12)
ax.hist(sample,bins=b,normed=1,alpha=0.3)
for q in mquantiles(sample,[0.025,0.5,0.975]):
    plt.axvline(q,linestyle='--')
ax.set_ylabel('P(x)',fontsize=12)
ax.set_xlabel('x',fontsize=12)
ax.set_title('CI for normal distribution')
fig.show()
low = mu-1.96*sigma
up = mu+1.96*sigma
print "Mean %4.2f,and CI are :%4.2f,%4.2f" %(mu, up,low) 
```

So our chance of observing a number less that -10.01 or greater that 9.99 are both less than 2.5%

# Central Limit Theorem (CLT); Conficence intervals and Statistical Testing

Central Limit Theorem states that if we repeatedly sample from any population (i.e. regardlss of the unerlying distribution of the measurement) and obtain a mean and a variance, after many sampling iterations the sampled mean will have a normal distribution with variance that is proportional to the population variance. Lets take a look visually what this means. The distribution on the left is clearly non symmetric, only positive and clearly skewed. The plot of the right is the estimation of the mean using repeated sampling from the distribution. This one IS normal as the CLT suggests. You can see that repeated sampling allows to estimate the mean of the  distribution. We can also estimate the population std which is given by:

$$\sigma   = \sigma_N \sqrt{N} $$



```python
%matplotlib inline
random.seed(10)
gammas = np.random.gamma(1,10,10000) # generate a 100,000 data points
fig,(ax0,ax1) = plt.subplots(1,2)
fig.set_figwidth(10)
fig.set_figheight(8)
ax0.hist(gammas,bins = np.arange(0,round(max(gammas))))
ax0.set_ylabel('count')
ax0.set_title('mean='+str(np.mean(gammas)))
ax0.set_xlabel('x')
clt = [np.mean(gammas[np.random.choice(10000,1000)]) for k in range(1000)] # sample and resample 1000 time a 1000 from the 100000.
ax1.hist(clt,bins =50)
ax1.set_title('mean='+str(np.mean(clt)))
ax1.set_xlabel('mean(x)')
#
print 'The distribution mean %4.2f is and the standared deviation is %4.2f ' %(np.mean(gammas),np.std(gammas))
print 'The sample CLI is mean %4.2f and the stanard error  is %4.2f' %(np.mean(clt),np.std(clt))
print 'The erimated population std is %4.3f' %(np.sqrt(1000)*np.std(clt))
```

With that in mind we are ready to build the confidence interval (CI) for the mean. The 95% CI is defined in the following way.
If we sample repeatedly from a distribution, 95% of the time the true mean will fall with in the limit of the CI.
This is precisely what we showed in the plot above. Graphically we observe 

```python

qq=mquantiles(clt,[0.025,0.5,0.975])
lowerr = qq[1]-qq[0]
upperr = qq[2]-qq[1]

plt.hist(clt,bins=50,normed=1,alpha=0.5)
plt.errorbar(qq[1],np.max(clt),xerr=[[lowerr],[upperr]],fmt='o',capsize=5, color='red')
for qnt in qq:
    plt.axvline(qnt,linestyle='--')
    
print 'The mean and 95 percent CLI are %4.2f (%4.2f,%4.2f)' %(qq[1], qq[0],qq[2])
```

## Comparing different population, t-testing p-value and all that jazz.

Motivating example I: Peripheral nerve blocks of the upper extremity will make for a faster recovery and discharge time. Lets 
compare the length of stay in PACU of general anesthesia VS peripheral nerve blocks. Typically the mean time in PACU is 15-20 minutes (i.e. 0.25-0.33 of an hour). Patient having a GA will stay on average an hour with a 30-40 minute standard deviation. Lets simulate the results and plot the CIs. We have:

```python
random.seed(10)
k1block=1
theta1block=0.33
g1block= np.random.gamma(k1block,theta1block,100000) # simulate times for PNC times in recovery

#
k0block=5
theta0block=0.2
g0block=np.random.gamma(k0block,theta0block,100000) # simulate time to Non PNC in recovery
print 'mean and standard deviation of peripheral nerve block are %4.4f and %4.4f respectively' %(np.mean(g1block),np.std(g1block))
print 'mean and standard deviation of NO peripheral nerve block are %4.4f and %4.4f respectively' %(np.mean(g0block),np.std(g0block))
#
# Get CLT for each one of the groups.
clt0block = [np.mean(g0block[np.random.choice(10000,1000)]) for k in range(1000)]
clt1block = [np.mean(g1block[np.random.choice(10000,1000)]) for k in range(1000)]
# Lets get the 95% CI for both groups
cli0block = mquantiles(clt0block,[0.025,0.5,0.975])
cli1block = mquantiles(clt1block,[0.025,0.5,0.975])

#
#Lets plot the CI we obtained. Is there a difference? in time spent in recovery
x = np.array([cli0block[1],cli1block[1]]) # median/mean estimates for mean 
y = np.array([1,1.2])

# error bar values w/ different -/+ errors
lower_error = x-np.array([cli0block[0],cli1block[0]])     # lower CI
upper_error = np.array([cli0block[2],cli1block[2]])-x     # upper CI
asymmetric_error = [lower_error, upper_error]

fig, ax1 = plt.subplots(nrows=1, sharex=True)


ax1.errorbar(x, y, xerr=asymmetric_error,  capsize=4,fmt='o')
ax1.set_title('Do we make a difference with block in delaying the room')
ax1.set_ylim([0.8,1.4])
ax1.set_yticks([1,1.2])
ax1.set_yticklabels(['no block','block'],fontsize=12)
ax1.set_xlabel('mean PACU times')
plt.show()

```

We observe from the plotting distribution that the PACU times are really generated from different 
processes and that the likelihood of observing 'block' PACU time in 'no block' is exceedingly low.
We can conclude that they probability of observing the above phenomenon under the assumption that the 
means for the the times in the groups are the same are less than 5% or 0.05. This is basically the meaning of p-value.

# Binomial Distribution

A frequency distribution of the possible number of successful outcomes in a given number of trials in each of which there is the same probability of success (p). The density function is given by 

$$P(k|p,n) = {{n}\choose{k} }p^k(1-p)^{n-k}$$

Motivating example I: Peripheral neuropahy after nerve blocks.  The probability of a peripheral nerve injury  from a nerve block is 0.0001, the risk from Tourniquet placement at 400mmHg over extended period of time is 0.07.
The Seattle VA does 100 nerve blocks a year So we have p1 is 0.0001 and the n=100. What is the probability of  observing 1,2,3,4,5.... 100 peripheral nerve injury?. Say in 1 year the VA regional anesthesia team has 10 peripehral nerve injuries, is that reasonable?.


```python
random.seed(123)
n=20
x=range(15)
p1=0.0001
p2=0.07
plt.plot(x,binom.pmf(x,n,p1),'bo',label='block',ms=5,color='blue')
plt.plot(x,binom.pmf(x,n,p2),'bo',label='tourniquet',ms=5,color='red')
plt.legend()
```

```python
n =20 # number of 'participants' or trials
N = 10000 # number draws
sim = np.random.binomial(1,p2,(N,n)) # in this case the trial size is 1/0
binsum = sim.sum(axis=1)
h = np.array([np.sum(binsum==k)for k in range(n)])
plt.plot(range(n),h/float(sum(h)),'bo',ms=5)
plt.grid('on')
```

# Part II - Logistic regression function,odds ratio, sensitivity/specificity estimation from data  and ROC curves.

Linear regression is a common way to model binary ourcomes and estimate the probabilities from binary data. It is given by 
$$logit(OR) = \frac{\log(p(x))}{1-\log(p(x))}=\beta_0 + \beta_i\mathbf{X_i}$$ where $OR$ is the Odds Ratio, $X_i$ is the features matrix, $p(x)$ is the probability for an outcome. Once we have the $OR$ calculating the probability is straight forward. The $\beta$ are coefficients for the straight line fitting.
We can see that calculating the $OR$ is given by:
$$OR  = \exp^{\beta_0 + \beta_i\mathbf{X_i}}$$
Lets look at an example relating to addmission data.
This methodlogy has been applied in multiple  settings. Any time we are interested in estimating the OR for a clinical conidtion logisitc regression is a straight forward way to estimate just that. i.e. smoking and Myocardial infarction in the Fremingham trail is an example.

## Sensitivity and specificity and its relation to performance of the classifier

We often speak of sensitivity and specificity of medical testing. This is basically addressing the quality of the tool/test/algorithm to classify the patient for example to healthy VS. disease state.
The Sensitivity of a test is defined as : $$P(T^+|D^+)$$
So the probability of a positive results GIVEN a trued state of diease.
Specificity is defined: $$P(T^-|D^-)$$
Put differently the probability of a negative test result given a true state of health.



We are often intrested in the performance of the classifier. Above we have fit a model to the training set and learned the model's
coefficienct. Next we wish to observe how well it performs. Remember we left 1/3 of the set out during training of the model. We now turn to the remining 1/3 of the dataset and define it as a test set. Remember these observatins were never used by the model. 
The assumption is that the same process we try to model that generated the training set also generated the test set. If so than our model should be able to predict the results of the test set. We can look at the predictions and estimate the error using the sensitivity and specificity using the Reciever Operator Characteristic (ROC) curve.

```python
# lets get the test set
test_set=df.drop(df.index[train_id])
y_test = test_set['admit']
x_test = test_set.ix[:,1:]
#
# Compute probabilities for being 0 or 1
scores = result.predict(x_test)
pred_df = pd.DataFrame()
pred_df['scores']=scores
pred_df.head()
```

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test, scores) # calulates
```

```python
fig,ax = plt.subplots()
fig.set_figwidth(12)
fig.set_figheight(10)
ax.plot(fpr,tpr,label='AUC=%4.2f' % metrics.auc(fpr,tpr))
ax.plot([0,1],[0,1],linestyle='--')
ax.set_ylabel('Sensitivity/TPR',fontsize=12)
ax.set_xlabel('1-spec/FPR',fontsize=12)
ax.set_title('ROC curve',fontsize=12)
ax.legend(loc=2)
ax.grid('on')

```

The above ROC has an average performance at best.  For instance we will achieve a sensitivity of 0.9 at a cost of at least 0.7 probabiity of a false positive rate. This is not a specific classifier. The ROC curve is bisected on the diagnonal to denote an area of  0.5 or a probability of 50%. Area of 0.5 is the area under the curve of using
a coin flipping as a classifier. We would typically hope to see  the ROC curve of our model above the diagonal and with a total area
close to 1.0.

Using VA databases, 35,000 Veterans were found to have TKA surgery in the VA. They were matched (on the correct operated side)  to all revisions and knee manipulation done in the VA. All opioids prescribed before surgery were abstracted from pharmacy records using text mining and were converted to Morphine Equivalent Dose.  The exact cause for revision was inferred from the post operative surgical note using methods of natural langauge processing and machine learning. Logistic regression and survival analysis were used to model the odds ration of having a revision or knee manipulation  in patients who were on long term therapy.
